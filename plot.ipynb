```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')

# Load CSV file
#ComputerVisionProject/SaMam/OUTPUT/ps8_lr2e-04_bs8_it50k_id1_es1_hb0_ssim0_ssim_w2_20260505_205008/loss_log.txt
#ComputerVisionProject/SaMam/OUTPUT/ps8_lr2e-04_bs8_it4k_id1_es0_hb1_ssim0_ssimW2_20260506_201459/loss_log.txt
# Replace 'data.csv' with your actual file path
#ComputerVisionProject/SaMam/OUTPUT/ps8_lr2e-04_bs8_it50k_id1_es1_hb1_ssim1_ssim_w2_20260508_015258/loss_log.txt
df = pd.read_csv('./OUTPUT/ps8_lr2e-04_bs8_it50k_id1_es1_hb1_ssim1_ssim_w2_20260508_015258/loss_log.txt')

# Inspect columns (optional, good for debugging)
print(df.head())

# Plot selected columns against 'step'
plt.figure()

plt.plot(df['step'], df['val_loss'], label='val_loss')
plt.plot(df['step'], df['content'], label='content')
plt.plot(df['step'], df['style'], label='style')
plt.plot(df['step'], df['id1'], label='id1')
plt.plot(df['step'], df['id2'], label='id2')
plt.plot(df['step'], df['ssim'], label='ssim')

# Labels and legend
plt.xlabel('Step')
plt.ylabel('Value')
plt.title('Training Metrics Over Steps')
plt.legend()

# Show plot
plt.show()
```
